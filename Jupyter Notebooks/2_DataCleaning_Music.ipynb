## Data Cleaning Web scapped Music Data

```python
#remove speech marks: 

topsongswiki_df['artist_song'] = list(map(lambda x: x.replace('"',''),topsongswiki_df['artist_song']))
topsongswiki_df
```

```python
#split songs to artist
topsongswiki_df[['artist','song']] = topsongswiki_df.artist_song.str.split(" – ", 1,expand=True,)
topsongswiki_df
```

```python
#drop artist_song
topsongswiki_df = topsongswiki_df.drop(['artist_song'], axis=1)
```

```python
topsongswiki_df.info()
```

```python
#remove duplicates
topsongswiki_df.drop_duplicates()
```

```python
# APPEND both data frames
topsongs_df = topsongswiki_df.append(topsongsBB_df)
```

```python
#append both datasets 
topsongs_df.append(topsongswiki_df)
```

```python
topsongs_df = topsongs_df.drop_duplicates()
```
